### 17. Write one line of code to compute the **average tip** amount of men who smoke and dine on **Friday's dinner**.


**Output should be a number**

````python
# Your answer/code here:
df[(df['sex'] == 'Male') & (df['smoker'] == 'Yes') & (df['day'] == 'Fri') & (df['time'] == 'Dinner')]['tip'].mean()
````

### 18. Write one line of code to compute the **sum** of total bills of the **male** customers who dine in any day that's **not Friday**.
