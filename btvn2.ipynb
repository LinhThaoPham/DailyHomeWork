### Task

The Age column's datatype is continuous. When analysing continuous column like this, there are oftentimes that we break them into categories. For example, people in 15 and 16 years old's behavior are not that much different so we group them into a category called "Teens".

Compute a new column called **AgeRange** in the dataset, following the criteria that:

- "Infants": if the passenger age < 1

- "Children": if age < 10

- "Teens": if age < 18

- "Adults": if age < 40

- "Middle Age": if age < 60

- "Elders": if age >= 60




```python
# HINT: Fill in the blanks ___

def cal_age_range(r):
    if r['Age'] < 1:
        return 'Infants'
    elif r['Age'] < 10:
        return 'Children'
    elif r['Age'] < 18:
        return 'Teens'
    elif r['Age'] < 40:
        return 'Adults'
    elif r['Age'] < 60:
        return 'Middle Age'
    elif r['Age'] >=60:
        return 'Elders'

df['AgeRange'] = df.apply(cal_age_range, axis = 1)
```
